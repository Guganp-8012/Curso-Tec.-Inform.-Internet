```python
# Decimal para binário

decimal = int(input("Digite um número decimal: "))
binario = ""

if decimal == 0:
    binario = "0"
else:
    while decimal > 0:
        binario = str(decimal % 2) + binario # resto = decimal % 2 -> str(resto)
        decimal = decimal // 2

print("O número em binário é:", binario)
```
